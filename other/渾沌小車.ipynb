```python
import numpy as np
import pandas as pd

def f(x):   
    xlst=[]
    for i in range(1000):
        xlst.append(x)
        x = 2 * x
        if x > 100:
            x = x - 100
        else:
            pass
    ser = pd.Series(xlst)
    return ser.value_counts().sort_index().shape[0]
```

```python
for x in range(101):
    print('%d : %d'%(x, f(x)))
```
