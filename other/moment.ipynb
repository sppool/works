```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
y = (np.random.choice(21, 80)) - 10
```

```python
rag = range(-10, 11)
```

```python
m = 0.9
lst = [random.choice(rag)]
for i in range(len(y)-1):
    val = random.choice(rag)
    new = (m * lst[-1])+((1-m) * val )
    lst.append(new)
lst = np.array(lst)
```

```python
plt.plot(y)
plt.plot(lst)
plt.ylim(20,-20)
```
